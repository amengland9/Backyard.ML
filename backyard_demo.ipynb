## Task 2: Pubchem Scrape and Parse SMILES

```python
def get_smiles_from_pubchem(compound_name):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    prop_url = f"{base_url}/compound/name/{compound_name}/property/CanonicalSMILES,ConnectivitySMILES/JSON"
    r = requests.get(prop_url)
    if r.status_code != 200:
        return None
    try:
        props = r.json()['PropertyTable']['Properties'][0]
        if 'CanonicalSMILES' in props:
            return props['CanonicalSMILES']
        elif 'ConnectivitySMILES' in props:
            return props['ConnectivitySMILES']
        else:
            return None
    except:
        return None
plant = input("Enter a plant name (case sensitive): ").strip().lower()
if plant in plant_to_compound:
    compound = plant_to_compound[plant]
    smiles = get_smiles_from_pubchem(compound)
    if smiles:
        print(f"\nRetrieved SMILES for {compound} (from {plant.title()}):\n{smiles}")
    else:
        print(f"Could not retrieve SMILES for {compound}.")
else:
    print(" Plant not in Greenhouse. Please check spelling or try another.")
```

## Task 3: Use the Model to Classify Plant of Choice

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import numpy as np
import sweetviz as sv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import Descriptors
from matplotlib.colors import ListedColormap
# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.preprocessing import StandardScaler
# Tree Visualisation
from IPython.display import Image

native = list(native_plants.keys())
nonnative = list(nonnative_plants.keys())
native_smiles = [get_smiles_from_pubchem(native_plants[plant]) for plant in native_plants]
nonnative_smiles = [get_smiles_from_pubchem(nonnative_plants[plant]) for plant in nonnative_plants]
total_smiles = native_smiles+nonnative_smiles
total_plants = native+nonnative
df = pd.DataFrame({'Plants':total_plants, 'SMILES':total_smiles})
mols = [Descriptors.CalcMolDescriptors(Chem.MolFromSmiles(mol)) for mol in df['SMILES']]
props = pd.DataFrame(mols)
props.replace(0, np.nan, inplace=True)
props.dropna(axis=1, how="all", inplace=True)
props.replace(np.nan, 0, inplace=True)
props["Plants"]= df["Plants"]
props["SMILES"]= df["SMILES"]
native_props_processed = props.iloc[:len(native_plants)]
native_props_processed = native_props_processed[["Plants", "SMILES","MaxAbsPartialCharge","MinAbsPartialCharge","MolLogP","MolMR"]]
native_props_processed["Native"] = 1
nonnative_props_processed = props.iloc[len(native_plants):]
nonnative_props_processed = nonnative_props_processed[["Plants", "SMILES","MaxAbsPartialCharge","MinAbsPartialCharge","MolLogP","MolMR"]]
nonnative_props_processed["Nonnative"] = 0
combined_data = pd.concat([native_props_processed, nonnative_props_processed], ignore_index=True)
combined_data.replace(np.nan, 0, inplace=True)
combined_data.drop_duplicates(inplace=True)
combined_data.reset_index(drop=True,inplace=True)
combined_data.to_csv("model-data.csv",index=False)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import Descriptors
from matplotlib.colors import ListedColormap
# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
# Tree Visualisation
from IPython.display import Image

combined = pd.read_csv("model-data.csv")
minCharge = combined["MinAbsPartialCharge"].to_list()
maxCharge = combined["MaxAbsPartialCharge"].to_list()
logP = combined["MolLogP"].to_list()
MR = combined["MolMR"].to_list()
native_dist = combined["Native"].value_counts()
print("Distribution of Native (1) vs. Native (0) Plants:")
print(native_dist)
native_percent = combined["Native"].value_counts(normalize=True) * 100
print("\nPercentage Distribution:")
print(native_percent)
plt.figure(figsize=(6,5))
sns.countplot(x="Native", data=combined, palette="viridis")
plt.title('Distribution of Native vs. Nonnative Plants')
plt.xlabel('Native (1) / Nonnative (0)')
plt.ylabel('Number of Plants')
plt.xticks([0, 1], ['Nonnative', 'Native'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
molecular_properties = [
    "MinAbsPartialCharge",
    "MaxAbsPartialCharge",
    "MolLogP",
    "MolMR"
]
for prop in molecular_properties:
    print(f"\n--- Property: {prop} ---")
    print(combined.groupby('Native')[prop].agg(['mean', 'std']))
    # Visualize the differences using box plots or violin plots
    plt.figure(figsize=(15, 10)) # Adjust figure size based on number of properties
    for i, prop in enumerate(molecular_properties):
        plt.subplot(2, (len(molecular_properties) + 1) // 2, i + 1) # Adjust subplot grid dynamically
        sns.boxplot(x='Native', y=prop, data=combined, palette='coolwarm')
        plt.title(f'{prop} by Nativity')
        plt.xlabel('Native (1) / Nonnative (0)')
        plt.ylabel(prop)
        plt.xticks([0, 1], ['Nonnative', 'Native'])
    plt.tight_layout()
    plt.show()
    
X = combined.drop(["Native", "Plants", "SMILES"], axis=1)
y = combined["Native"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
param_dist = {"n_estimators": randint(50,500),
             "max_depth": randint(1,20)}
rf = RandomForestClassifier()
rand_search = RandomizedSearchCV(rf,
                                param_distributions = param_dist,
                                n_iter=5,
                                cv=5)
rand_search.fit(X_train, y_train)
best_rf = rand_search.best_estimator_
print('Best hyperparamaters:', rand_search.best_params_)
y_pred = best_rf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False)
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted Native')
plt.ylabel('True Native')
plt.show()

final_predict = pd.DataFrame({'Predicted': best_rf.predict(X)})
final_df = pd.concat([combined, final_predict], axis=1)
print(final_df)

def get_native_from_smiles(smiles, final_df):
    final_dict = dict(zip(final_df["SMILES"].to_list(),final_df["Predicted"]))
    return final_dict.get(smiles) > 0

compound = plant_to_compound[plant]
smiles = get_smiles_from_pubchem(compound)
get_smiles_from_pubchem()
```
