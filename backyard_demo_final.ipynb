## Task 2: Pubchem Scrape and Parse SMILES

```python
def get_smiles_from_pubchem(compound_name):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    prop_url = f"{base_url}/compound/name/{compound_name}/property/CanonicalSMILES,ConnectivitySMILES/JSON"
    r = requests.get(prop_url)
    if r.status_code != 200:
        return None
    try:
        props = r.json()['PropertyTable']['Properties'][0]
        if 'CanonicalSMILES' in props:
            return props['CanonicalSMILES']
        elif 'ConnectivitySMILES' in props:
            return props['ConnectivitySMILES']
        else:
            return None
    except:
        return None
```
